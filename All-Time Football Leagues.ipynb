```python
# Step 1: Fetch the HTML content of the webpage
url1 = 'https://www.worldfootball.net/alltime_table/eng-premier-league/'
response = requests.get(url1)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Step 2: Locate the table in the HTML
    table1 = soup.find('table', {'class': 'standard_tabelle'})
    
    # Step 3: Extract headers
    headers1 = [th.text.strip() for th in table1.find_all('th')]
    print(f"Headers (Columns): {headers1}")
    print(f"Number of Columns in Headers: {len(headers1)}")
    
    # Step 4: Extract table rows
    rows = []
    for tr in table1.find_all('tr')[1:]:  # Skip header row
        cols = [td.text.strip() for td in tr.find_all('td')]
        if cols:  # Skip empty rows
            rows.append(cols)
    
    print(f"First Row: {rows[0]}")
    print(f"Number of Columns in First Row: {len(rows[0])}")

    # Step 5: Ensure Column Alignment
    if len(headers1) < len(rows[0]):
        # Add placeholder headers for extra columns
        headers1 += [f"Extra Column {i+1}" for i in range(len(rows[0]) - len(headers1))]
    elif len(headers1) > len(rows[0]):
        print("Warning: Headers have more columns than data rows.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

df_eng = pd.DataFrame(rows, columns = headers1)
```

# Making a Win Probability Model

### Using Linear Regression, Measuring a Team's Probability of Winning based on Goals Scored, Goals Conceded, and Matches Played so far
### Also Measuring each Variable's Quantified Correlation with the Outcome of "Win"

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from sklearn.preprocessing import StandardScaler as SS
```

```python
# Defining the Features (X) and Target/Outcome variable (y) for Multiple Linear Regression
X = df[["WPG", "DPG", "LPG"]]
y = df["PPG"]

# Normalizing the Features & the Target variable
scaler = SS()

X_norm = scaler.fit_transform(X)
y_norm = scaler.fit_transform(y.values.reshape(-1, 1))

# Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size = 0.2, random_state=42)

# Creating and Fitting the Multiple Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Coefficients and Y-Intercept
intercept = model.intercept_
coefficients = model.coef_

print(intercept)
print("")

coef_df = pd.DataFrame({"Feature": ["Wins PG", "Draws PG", "Losses PG"], "Coefficient": coefficients.flatten()})
print(coef_df) # Coefficients taken from the "coefficients" DataFrame created
```

```python
# Making Predictions on the Test Set
y_pred = model.predict(X_test)
```
