```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL of the stats page
url = "https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;template=results;type=bowling"

# Create a session with a User-Agent header
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find all tables
    tables = pd.read_html(str(soup))
    
    # Display the number of tables found
    print(f"Number of tables found: {len(tables)}")
    
    # Extract the first table
    if tables:
        df = tables[0]
        print(df)
        # Optionally save the DataFrame
        df.to_csv("espn_cricinfo_table.csv", index=False)
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python
# We Need to Ensure that a Bowler Has Bowled at least 5000 Deliveries, to get a Good-Enough Sample Size
bowsam = bow_df[bow_df["Balls"] > 5000]
```

```python
plt.figure(figsize=(7, 8))

sns.boxplot(data = bowsam, y="Econ")
plt.ylabel("Bowling Economy")

median = bowsam["Econ"].median()
average = bowsam["Econ"].mean()
q1 = bowsam["Econ"].quantile(0.25)
q3 = bowsam["Econ"].quantile(0.75)

plt.axhline(median, color='blue', linestyle='--', label=f'Median: {median:.2f}')
plt.axhline(average, color='orange', linestyle='--', label=f'Average: {average: .2f}')
plt.axhline(q1, color='black', linestyle='--', label=f"1st Quartile: {q1: .2f}")
plt.axhline(q3, color = 'green', linestyle='--', label=f"3rd Quartile: {q3: .2f}")

plt.legend(loc='upper right')

plt.show()
```
